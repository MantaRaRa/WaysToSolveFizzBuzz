FizzBuzz printing numbers from 0 to 100

```python
for num in range(1,101):
    print(num)
```

variation: FizzBuzz printing numbers from 0 to 100

```python
for num in range(101):
    print(num)
```

Fizzing the numbers that are divisible by 3

```python
for num in range(101):
    
    if num % 3 == 0:
        print("Fizz")
    else:
        print(num)
```

Buzzing the numbers that are divisible by 5

```python
for num in range(101):

    if num % 5 == 0:
        print("Buzz")    
    else:
        print(num)
```

FizzBuzzing

```python
for num in range(101):
    if num % 3 == 0 and num % 5 == 0:
        print("fizzbuzz")
        continue
    elif num % 3 == 0:
        print("fizz")
        continue
    elif num % 5 == 0:
        print("buzz")
        continue
    print(num)
```

FizzBuzz STRING CONCATENATION

```python
for num in range(101):
    string = ""
    if num % 3 == 0:
        string = string + "Fizz"
    if num % 5 == 0:
        string = string + "Buzz"
    if num % 5 != 0 and num % 3 != 0:
        string = string + str(num)
    print(string)
```

FizzBuzz LAMBDA

```python
print(*map(lambda i: 'Fizz'*(not i%3)+'Buzz'*(not i%5) or i, range(101)),sep='\n')
```
